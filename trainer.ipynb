```python
# TRAINER

import numpy as np

theta0 = 0.0;
theta1 = 0.0;

tmpTheta0 = 0.0;
tmpTheta1 = 0.0;

mileage_list = None;
price_list = None;


sum_errors = 0.0;

min_mileage_val = 0.0;
max_mileage_val = 0.0;
min_price_val = 0.0;
max_price_val = 0.0;

with open("theta_settings.txt", "w") as file:
    file.write(f"{theta0}\n");
    file.write(f"{theta1}");


learningRate = 0.01; 
num_iterations = 100000 # total number of iterations to train our ai


# numpy way to split strings into arrays, fast
# delimiter=',' tells it to split at the comma
# unpack=True splits the columns into separate arrays
# skiprows=1 : skip the first line
data = np.loadtxt("data.csv", delimiter=",", skiprows=1);
mileage_list = data[:,0];		
price_list = data[:,1]; 	


# Normalization of the mileage values, prevents overflows
min_mileage_val = mileage_list.min();
max_mileage_val = mileage_list.max();

mileage_normalized = (mileage_list - min_mileage_val) / (max_mileage_val - min_mileage_val);


# Normalization of the price values
min_price_val = price_list.min();
max_price_val = price_list.max();

price_normalized = (price_list - min_price_val) / (max_price_val - min_price_val);

m = mileage_list.size;	# total number of elements


for i in range(num_iterations):
# --- ONE STEP OF GRADIENT DESCENT / The formulas ---
	sum_errors = 0.0;
	sum_errors_mileage = 0.0;

	# Calculate errors for every car
	for j in range(m):
		estimatedPrice = theta0 + (theta1 * mileage_normalized[j]);
		error = estimatedPrice - price_normalized[j];
		
		sum_errors += error;
		sum_errors_mileage += (error * mileage_normalized[j]);

	# Calculate the new values using the formulas
	tmpTheta0 = theta0 - learningRate * (sum_errors / m);
	tmpTheta1 = theta1 - learningRate * (sum_errors_mileage / m);

	# Update thetas
	theta0 = tmpTheta0;
	theta1 = tmpTheta1;


# Finally, write thetas values for the predictor
with open("theta_settings.txt", "w") as file:
    file.write(f"{theta0}\n");
    file.write(f"{theta1}");

# Write the normalization values for the predictor
with open("normalize_settings.txt", "w") as file:
    file.write(f"{min_mileage_val}\n");
    file.write(f"{max_mileage_val}\n");
    file.write(f"{min_price_val}\n");
    file.write(f"{max_price_val}\n");

print("Model trained. You can now use the predictor.\n");
```
