```python
# PREDICTOR

theta0 = 0.0
theta1 = 0.0

min_mileage_val = 0.0;
max_mileage_val = 0.0;
min_price_val = 0.0;
max_price_val = 0.0;

# Get saved thetas
with open("theta_settings.txt", "r") as file:
	# .strip() removes the "\n"
	# float() converts the string to float
	theta0 = float(file.readline().strip());
	theta1 = float(file.readline().strip());


# Get normalization settings
with open("normalize_settings.txt", "r") as file:
	min_mileage_val = float(file.readline().strip());
	max_mileage_val = float(file.readline().strip());
	min_price_val = float(file.readline().strip());
	max_price_val = float(file.readline().strip());

mileage_input = float(input("Enter the mileage to predict: "));

# Normalize the input
normalized_mileage_input = (mileage_input - min_mileage_val) / (max_mileage_val - min_mileage_val);

# Normalize the result
estimatedPrice_normalized = theta0 + (theta1 * normalized_mileage_input);

# Denormalize result
estimatedPrice = estimatedPrice_normalized * (max_price_val - min_price_val) + min_price_val;


print(f"Estimated price for mileage {mileage_input} is {estimatedPrice}");
```
